```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data2020 = pd.read_excel('../Data/Векторизированные2020.xlsx').drop('Unnamed: 0',axis=1)
data2021 = pd.read_excel('../Data/Векторизированные2021.xlsx').drop('Unnamed: 0',axis=1)
data2022 = pd.read_excel('../Data/Векторизированные2022.xlsx').drop('Unnamed: 0',axis=1)
```

```python
id = list(data2020['ID Клиента'].unique())
```

```python
data2020[data2020['ID Клиента'] == 10406690.0]
```

```python
visits = {}

for i in id:
    visits['Client: {}'.format(i)] = len(data2020[data2020['ID Клиента'] == i])
```

```python
visits.pop('Client: nan')
```

```python
def discountCoef(id):
    discount = int(list(data2020[data2020['ID Клиента'] == id]['Предоставленная скидка'].values)[0])
    if discount != 0:
        coef = discount / 100
    else:
        coef = data2020['Предоставленная скидка'].median() / 100
    return coef
```

```python
discountCoef(10356763.0)
```

```python
coefficient = {}

for key,val in visits.items():
    id = float(key.split()[1])
    discoef = discountCoef(id)
    coefficient[id] = val/discoef
    
coefficient
```
