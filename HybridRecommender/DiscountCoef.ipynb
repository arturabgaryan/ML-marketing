```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import time
```

```python
data2020 = pd.read_excel('../Data/Векторизированные2020.xlsx').drop('Unnamed: 0',axis=1).dropna()
data2021 = pd.read_excel('../Data/Векторизированные2021.xlsx').drop('Unnamed: 0',axis=1)
data2022 = pd.read_excel('../Data/Векторизированные2022.xlsx').drop('Unnamed: 0',axis=1)
```

```python
id1 = list(data2020['ID Клиента'].unique())
```

```python
id2 = list(data2021['ID Клиента'].unique())
```

```python
id3 = list(data2022['ID Клиента'].unique())
```

```python
def visitsFun(dataset,ids):
    visits = {}
    count = 0
    for i in ids:
        visits[str(i)] = len(dataset[dataset['ID Клиента'] == i])
        count  += 1
    return visits

visits2020 = visitsFun(data2020,id1)
visits2021 = visitsFun(data2021,id2)
visits2022 = visitsFun(data2022,id3)
```

```python
def discountCoef(id,dataset):
    discount = int(list(dataset[dataset['ID Клиента'] == id]['Предоставленная скидка'].values)[0])
    if discount != 0:
        coef = discount / 100
    else:
        coef = dataset['Предоставленная скидка'].median() / 100
    return coef

discountCoef(10356763.0,data2020)
```

```python
def standart(coef):
    coef = (coef - coef.mean())/coef.std()
```

```python
dataset2020 = pd.DataFrame({'index': visits2020.values(),'id': visits2020.keys()},index=visits2020.keys())
dataset2021 = pd.DataFrame({'index': visits2021.values(),'id': visits2021.keys()},index=visits2021.keys())
dataset2022 = pd.DataFrame({'index': visits2022.values(),'id': visits2022.keys()},index=visits2022.keys())
```

```python
dataset2020['index'] = abs((dataset2020['index'] - dataset2020['index'].mean())/dataset2020['index'].std())
dataset2021['index'] = abs((dataset2021['index'] - dataset2021['index'].mean())/dataset2021['index'].std())
dataset2022['index'] = abs((dataset2022['index'] - dataset2022['index'].mean())/dataset2022['index'].std())
```

```python
dataMoney2020 = abs((data2020['Сумма расходов'] - data2020['Сумма расходов'].mean())/data2020['Сумма расходов'].std())
dataMoney2021 = abs((data2021['Сумма расходов'] - data2021['Сумма расходов'].mean())/data2021['Сумма расходов'].std())
dataMoney2022 = abs((data2022['Сумма расходов'] - data2022['Сумма расходов'].mean())/data2022['Сумма расходов'].std())
```

```python
coefficient = {}
coefficient2020 = {}
coefficient2020 = {}

def Coefficient(data,dataset,datamoney):
    coefficient = {}
    for i in range(len(data)):
        id = data.loc[i]['ID Клиента']
        discoef = discountCoef(id,data)
        visit = round(dataset[dataset['id'] == str(id)]['index'].values[0],5)
        money = float(datamoney.loc[i])
        coefficient[id] = (money*visit)*discoef
    return coefficient
    
coefficient2020 = Coefficient(data2020,dataset2020,dataMoney2020)
coefficient2021 = Coefficient(data2021,dataset2021,dataMoney2021)
coefficient2022 = Coefficient(data2022,dataset2022,dataMoney2022)
```

```python
coefficient2022 = dict(sorted(coefficient2022.items(),key = lambda item:item[1]))

coefficient2022
```
