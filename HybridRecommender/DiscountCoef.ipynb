```python
def visitsFun():
    visits = {}
    count = 0
    for i in id:
        visits[str(i)] = len(data2020[data2020['ID Клиента'] == i])
        count  += 1
    return visits

visits = visitsFun()
```

```python
def discountCoef(id):
    discount = int(list(data2020[data2020['ID Клиента'] == id]['Предоставленная скидка'].values)[0])
    if discount != 0:
        coef = discount / 100
    else:
        coef = data2020['Предоставленная скидка'].median() / 100
    return coef

discountCoef(10356763.0)
```

```python
def standart(coef):
    coef = (coef - coef.mean())/coef.std()
```

```python
dataset = pd.DataFrame({'index': visits.values(),'id': visits.keys()},index=visits.keys())
```

```python
dataset['index'] = abs((dataset['index'] - dataset['index'].mean())/dataset['index'].std())
```

```python
coefficient = {}

for i in range(len(data2020)):
    id = data2020.loc[i]['ID Клиента']
    discoef = discountCoef(id)
    visit = dataset[dataset['id'] == id]['index']
    money = dataMoney.loc[i].value
    print(money,visit,discoef)
    
    coefficient[id] = money*visit/discoef
    
coefficient
```
