```python
import pandas as pd
```

```python
data2020 = pd.read_excel('../Data/Датасет 2020.xlsx').drop(['Год выдачи паспорта'],axis = 1)
data2021 = pd.read_excel('../Data/Датасет 2021.xlsx').drop(['Год выдачи паспорта'],axis = 1)
data2022 = pd.read_excel('../Data/Датасет 2022.xlsx').drop(['Год выдачи паспорта'],axis = 1)
```

```python
data2020 = data2020.replace(to_replace='скидки не было', value=0, regex=True)
data2021 = data2021.replace(to_replace='скидки не было', value=0, regex=True)
data2022 = data2022.replace(to_replace='скидки не было', value=0, regex=True)
```

```python
data2021.head()
```

```python
regions2020 = sorted(data2020['Регион проживания'].unique())
regions2021 = sorted(data2021['Регион проживания'].unique())
regions2022 = sorted(data2022['Регион проживания'].unique())
```

```python
print(regions2020,regions2021,regions2022,sep='\n\n\n')
```

```python
stores2020 = sorted(data2020['МАГАЗИН'].unique())
stores2021 = sorted(data2021['МАГАЗИН'].unique())
stores2022 = sorted(data2022['МАГАЗИН'].unique())
```

```python
print(stores2020,stores2021,stores2022,sep='\n\n\n')
```
