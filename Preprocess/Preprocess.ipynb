```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data2020 = pd.read_excel('../Data/Датасет 2020.xlsx').drop(['Год выдачи паспорта'],axis = 1)
data2021 = pd.read_excel('../Data/Датасет 2021.xlsx').drop(['Год выдачи паспорта'],axis = 1)
data2022 = pd.read_excel('../Data/Датасет 2022.xlsx').drop(['Год выдачи паспорта'],axis = 1)
```

```python
data2020 = data2020.replace(to_replace='скидки не было', value=0, regex=True)
data2021 = data2021.replace(to_replace='скидки не было', value=0, regex=True)
data2022 = data2022.replace(to_replace='скидки не было', value=0, regex=True)

data2020['Сумма расходов'] = data2020['Сумма расходов'].astype(int)
data2021['Сумма расходов'] = data2021['Сумма расходов'].astype(int)
data2022['Сумма расходов'] = data2022['Сумма расходов'].astype(int)
```

```python
def percentage(a):
    return 1 - int(a)/100
```

```python
data2022['Цена без скидки'] = (data2022['Сумма расходов']/(data2022['Предоставленная скидка'].apply(percentage))).astype(int)
data2021['Цена без скидки'] = (data2021['Сумма расходов']/(data2021['Предоставленная скидка'].apply(percentage))).astype(int)
data2020['Цена без скидки'] = (data2020['Сумма расходов']/(data2020['Предоставленная скидка'].apply(percentage))).astype(int)
```

```python
print(sorted(data2020['Возраст клиента'].unique()))
print(sorted(data2021['Возраст клиента'].unique()))
print(sorted(data2022['Возраст клиента'].unique()))
```

```python
dataBelow25 = pd.concat([data2020[data2020['Возраст клиента'] <= 25],
                         data2021[data2021['Возраст клиента'] <= 25],
                         data2022[data2022['Возраст клиента'] <= 25]],axis = 0)
```

```python
dataAbove35 = pd.concat([data2020[35 <= data2020['Возраст клиента']],
                         data2021[35 <= data2021['Возраст клиента']],
                         data2022[35 <= data2022['Возраст клиента']]],axis = 0)
```

```python
dataFrom26To34 = pd.concat([data2020[(25 < data2020['Возраст клиента']) & (data2020['Возраст клиента'] < 35)],
                            data2021[(25 < data2021['Возраст клиента']) & (data2021['Возраст клиента'] < 35)],
                            data2022[(25 < data2022['Возраст клиента']) & (data2022['Возраст клиента'] < 35)]],axis = 0)
```

```python
dataBelow25.to_excel('../Data/Датасет до 25.xlsx')
dataFrom26To34.to_excel('../Data/Датасет 25-34.xlsx')
dataAbove35.to_excel('../Data/Датасет от 35.xlsx')
```

```python
age_groups = []
age_groups.append(dict(dataBelow25['МАГАЗИН'].value_counts()))
age_groups.append(dict(dataFrom26To34['МАГАЗИН'].value_counts()))
age_groups.append(dict(dataAbove35['МАГАЗИН'].value_counts()))

keys_to_delete = []
for i in age_groups:
    for key,val in i.items():
        if val < 500:
            keys_to_delete.append(key)
    for j in keys_to_delete:
        i.pop(j)
    keys_to_delete = []
        
fig,ax = plt.subplots(3,1,figsize = (23,23))

sizes = list(age_groups[0].values())
labels = list(age_groups[0].keys())

ax[0].set_title("Group below age 25:")
ax[0].pie(sizes,labels=labels,autopct='%1.1f%%')

sizes = list(age_groups[1].values())
labels = list(age_groups[1].keys())

ax[1].set_title("Group between age 25-34:")
ax[1].pie(sizes,labels=labels,autopct='%1.1f%%') 

sizes = list(age_groups[2].values())
labels = list(age_groups[2].keys())

ax[2].set_title("Group above age 35:")
ax[2].pie(sizes,labels=labels,autopct='%1.1f%%') 

count += 1 
```
