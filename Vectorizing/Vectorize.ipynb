```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
data2020 = pd.read_excel('../Data/Датасет 2020.xlsx').drop(['Год выдачи паспорта'],axis = 1)
data2021 = pd.read_excel('../Data/Датасет 2021.xlsx').drop(['Год выдачи паспорта'],axis = 1)
data2022 = pd.read_excel('../Data/Датасет 2022.xlsx').drop(['Год выдачи паспорта'],axis = 1)
```

```python
data2020 = data2020.replace(to_replace='скидки не было', value=0, regex=True)
data2021 = data2021.replace(to_replace='скидки не было', value=0, regex=True)
data2022 = data2022.replace(to_replace='скидки не было', value=0, regex=True)

data2020['Сумма расходов'] = data2020['Сумма расходов'].astype(int)
data2021['Сумма расходов'] = data2021['Сумма расходов'].astype(int)
data2022['Сумма расходов'] = data2022['Сумма расходов'].astype(int)
```

```python
data2020 = data2020[data2020['Регион проживания'] != '-']
data2021 = data2021[data2021['Регион проживания'] != '-']
data2022 = data2022[data2022['Регион проживания'] != '-']
```

```python
regions2020 = sorted(data2020['Регион проживания'].unique())
regions2021 = sorted(data2021['Регион проживания'].unique())
regions2022 = sorted(data2022['Регион проживания'].unique())
```

```python
def BoW(feature):
    BoW = []
    for i in feature:
        sp = i.split()
        for z in sp:
            BoW.append(z)
    return BoW
            
def vectorize(feature,bow):
    vectors = {}
    for i in feature:
        vec = []
        sp = i.split()
        print(sp)
        for z in bow:
            if z in sp:
                vec.append(1)
            else:
                vec.append(0)
        vectors[i] = vec
        print(vec)        
        print('------------------'*5)
    return vectors
```

```python
Bow_stores2020 = BoW(data2020['МАГАЗИН'].unique())
stores2020 = vectorize(data2020['МАГАЗИН'].unique(),Bow_stores2020)

Bow_stores2021 = BoW(data2021['МАГАЗИН'].unique())
stores2021 = vectorize(data2021['МАГАЗИН'].unique(),Bow_stores2021)

Bow_stores2022 = BoW(data2022['МАГАЗИН'].unique())
stores2022 = vectorize(data2022['МАГАЗИН'].unique(),Bow_stores2022)
```

```python
Bow_reg2020 = BoW(data2020['Регион проживания'].unique())
regions2020 = vectorize(data2020['Регион проживания'].unique(),Bow_reg2020)

Bow_reg2021 = BoW(data2021['Регион проживания'].unique())
regions2021 = vectorize(data2021['Регион проживания'].unique(),Bow_reg2021)

Bow_reg2022 = BoW(data2022['Регион проживания'].unique())
regions2022 = vectorize(data2022['Регион проживания'].unique(),Bow_reg2022)
```

```python
Bow_cards2020 = BoW(data2020['Тип карточки (B-VISA, S-MASTERCARD или МИР)'].unique())
cards2020 = vectorize(data2020['Тип карточки (B-VISA, S-MASTERCARD или МИР)'].unique(),Bow_cards2020)

Bow_cards2021 = BoW(data2021['Тип карточки (B-VISA, S-MASTERCARD или МИР)'].unique())
cards2021 = vectorize(data2021['Тип карточки (B-VISA, S-MASTERCARD или МИР)'].unique(),Bow_cards2021)

Bow_cards2022 = BoW(data2022['Тип карточки (B-VISA, S-MASTERCARD или МИР)'].unique())
cards2022 = vectorize(data2022['Тип карточки (B-VISA, S-MASTERCARD или МИР)'].unique(),Bow_cards2022)
```

```python
def create_vec(n):
    Vector = []
    data2022.iloc[n]
    Vector.append(regions2022[data2022.iloc[n]['Регион проживания']])
    Vector.append(stores2022[data2022.iloc[n]['МАГАЗИН']])
    Vector.append(cards2022[data2022.iloc[n]['Тип карточки (B-VISA, S-MASTERCARD или МИР)']])
    Vector.append(data2022.iloc[n]['Сумма расходов'])
    Vector.append(data2022.iloc[n]['Предоставленная скидка'])
    Vector.append(data2022.iloc[n]['Возраст клиента'])
    return Vector

vecs = []
for i in range(len(data2022)):
    vecs.append(list(create_vec(i)))
```

```python
d = {'Вектор':vecs}
```

```python
dataplusvec2022 = pd.DataFrame(data = d)
```

```python
vectorized2022 = pd.concat([data2022,dataplusvec2022],axis = 1)
```
